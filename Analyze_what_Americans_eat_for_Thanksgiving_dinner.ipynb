# 5: Converting Age To Numeric

```python
data['Age']
```

```python
def convert(string):
    if pd.isnull(string):
        return None
    else:
        return int(string.split("-")[0].replace("+",""))
```

```python
data['int_age'] = data['Age'].apply(convert)
data.head()
```

```python
data['int_age'].describe()
```

#### The miniumum age is 18 and the maximum is 60, standard deviation is 15, 2nd quantile is 45

# 6: Converting Income To Numeric

```python
data['How_much_total_combined_money_did_all_members_of_your_HOUSEHOLD_earn_last_year?']
```

```python
def conversion(string):
    if pd.isnull(string):
        return None
    else:
        result = string.split(" ")[0]
        if result == 'Prefer':
            return None
        else:
            result = result.replace('$','')
            result = result.replace(',','')
            return int(result)
        
```

```python
data['int_income'] = data['How_much_total_combined_money_did_all_members_of_your_HOUSEHOLD_earn_last_year?'].apply(conversion)
```
